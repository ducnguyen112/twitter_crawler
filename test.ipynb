```python
import pandas as pd
```

```python
test_df = pd.read_csv('persons_202412310959.csv')
```

```python
import re
def check_company_like_name(row):
    name = row['name']
    company = row['company_name']
    return name.lower() in company.lower()
```

```python
test_df['company_like_name'] = test_df.apply(check_company_like_name, axis=1)
```
